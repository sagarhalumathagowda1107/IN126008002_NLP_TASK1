Task 1: Conceptual Understanding

1. "Love" vs "love" in NLP

In NLP, "Love" and "love" are treated as different words if we don’t convert everything to lowercase. That means the model might think they are unrelated even though they mean the same thing. That’s why most preprocessing steps convert text to lowercase to keep things consistent.

2. What happens if stopwords are not removed

If stopwords are not removed, the text will contain a lot of unnecessary words like "is", "the", "and". These don’t add much meaning but increase the size of the data. This can make the model slower and sometimes less accurate.

3. When removing stopwords can be harmful

Sentiment Analysis:
In a sentence like "not good", if we remove "not", it becomes "good", which completely changes the meaning.

Chatbots / Question Systems:
In a sentence like "what is your name", removing words like "what" or "is" can break the structure and make it harder for the system to understand the question.

4. Stemming vs Lemmatization

Stemming:
It simply cuts words to their base form without checking if it’s a real word. For example, "running" might become "runn".

Lemmatization:
It converts words to their correct root form using meaning and grammar. For example, "running" becomes "run", which is a proper word.

```python
import re

def preprocess_text(text):

    # 1. lowercase
    text = text.lower()

    # 2. remove urls
    text = re.sub(r'http\S+|www\S+', '', text)

    # 3. remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # 4. remove numbers
    text = re.sub(r'\d+', '', text)

    # 5. handle repeated characters (FINAL WORKING)
    text = re.sub(r'(.)\1+', r'\1\1', text)   # keep max 2 letters
    text = re.sub(r'\bso+\b', 'so', text)     # fix "sooo" → "so"

    # 6. remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # 7. remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 8. remove short words
    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ['i', 'is', 'no', 'not', 'so']]

    text = " ".join(words)

    return text
texts = [
    "I have 2 dogs!!!",
    "This is good",
    "soooo goooood!!!",
    "WOW!!! This is GREAT!!!",
    "Visit http://example.com now"
]

for t in texts:
    print(preprocess_text(t))
```

```python
texts = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Apply preprocessing
for t in texts:
    cleaned = preprocess_text(t)
    tokens = cleaned.split()

    print("Original Text :", t)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-" * 50)
```

```python
def token_analysis(text):
    tokens = text.split()

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens == 0:
        avg_length = 0
    else:
        avg_length = sum(len(word) for word in tokens) / total_tokens

    return total_tokens, unique_tokens, avg_length


# Use same cleaned texts
for t in texts:
    cleaned = preprocess_text(t)

    total, unique, avg = token_analysis(cleaned)

    print("Sentence :", cleaned)
    print("Total tokens :", total)
    print("Unique tokens :", unique)
    print("Average token length :", round(avg, 2))
    print()
```

Analysis Questions
1. Which sentence had the most noise?
The sentence "Get 100% FREE access now!!!" had the most noise because it contained numbers, uppercase letters, and special characters. These elements do not contribute meaningful information and were removed during preprocessing.

2. Which sentence retained the most meaningful tokens after cleaning?
The sentence "I absolutely loooved this product" retained the most meaningful tokens after cleaning because it contains descriptive and sentiment-rich words like "absolutely", "loved", and "product", which provide clear context and meaning.

```python
from collections import Counter

# Step 1: Collect all tokens
all_tokens = []

for t in texts:
    cleaned = preprocess_text(t)
    tokens = cleaned.split()
    all_tokens.extend(tokens)

# Step 2: Count frequency
word_freq = Counter(all_tokens)

# Step 3: Top 10 most frequent words
top_10 = word_freq.most_common(10)

# Step 4: Top 5 least frequent words
least_5 = sorted(word_freq.items(), key=lambda x: x[1])[:5]

# Output
print("Top 10 Most Frequent Words:")
for word, count in top_10:
    print(f"{word} : {count}")

print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(f"{word} : {count}")
```

```python
def full_pipeline(text_list):

   all_tokens = []
   clean_sentences = []

   for text in text_list:
      cleaned = preprocess_text(text)
      tokens = cleaned.split()

      clean_sentences.append(cleaned)
      all_tokens.extend(tokens)

   return {
    "tokens": all_tokens,
    "clean_sentences": clean_sentences
}

result = full_pipeline(texts)

print("Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
print(result["clean_sentences"])
```
